Além do exemplo da aula passada, vamos tentar fazer um backtesting direto do DataFrame, também chamado de Backtesting Vetorial:

```python
df = pd.read_csv('^BVSP.csv').dropna()
df
```

Calcular os retornos:

```python
dfr = df.Close.pct_change(1).dropna()
dfr
```

Gerar um sinal:

```python
# Se o retorno atual for maior que zero, compra, senão vende
# Aqui entra o modelo de sinal

signal = (dfr > 0)*2-1
signal
```

```python
# Apuração de resultado.
# Cuidado com o look-ahead bias

result = signal.shift(1) * dfr
result
```

```python
# Resultado Total:

total = result.sum()
total
```

```python
# Resultado Acumulado

result.cumsum().plot();
```

Para fazer exemplos mais complexos, precisamos de outro framework. O backtesting vetorial é bem limitado e permite apenas modelos de sinal mais simplificados.

Vamos começar importando a biblioteca para simular o Backtesting (event-driven):

```python
from backtesting import evaluateHist, evaluateIntr, Strategy, Order
```

Agora vamos fazer uma primeira simulação com o classico Buy & Hold

```python
# Criando sua estratégia
class BuynHold(Strategy):

    # Chamado de construtor: inicializa variáveis de suporte
    def __init__(self):
        self.bought = False # Indica se já comprou ou não

    # Função chamada a cada preço recebido
    def receive(self, event):
        
        # Se ainda não comprou, faça
        # Note o uso do prefixo self. para as variáveis de suporte
        if not self.bought:
            self.bought = True
            
            # Envia uma ordem compra de 1 a mercado (preço zero)
            self.submit(self._id, Order(event.instrument, Order.B, 1, 0))
    
    # Função chamada a cada ordem preenchida
    def fill(self, id, instrument, price, quantity, status):
        print(f'Order Filled: {instrument} {quantity}@{price}')

# Rodando a simulação com dados históricos do IBOV
# Formato: evaluateHist(MyStrategyName(), {'instumento': 'nome do arquivo'})

print(evaluateHist(BuynHold(), {'IBOV': '^BVSP.csv'}))
```

Outro exemplo com mais elementos:

```python
class MonkeyTrade(Strategy):
    
    def __init__(self):
        self.side = 0 # Qual lado está posicionado

    def receive(self, event):

        if random.randint(1, 10) == 1: # se sair 1 de um sorteio de 1 a 10
            side = random.choice([-1, 1]) # sorteia um lado: -1 ou 1
            if side != self.side: # se o lado está diferente
                
                # Zera a posição atual
                if self.side == 1:
                    self.submit(self._id, Order(event.instrument, Order.S, self.side, 0))
                elif self.side == -1:
                    self.submit(self._id, Order(event.instrument, Order.B, -self.side, 0))
                    
                # Monta a posição nova
                if side == 1:
                    self.submit(self._id, Order(event.instrument, Order.B, side, 0))
                else:
                    self.submit(self._id, Order(event.instrument, Order.SS, -side, 0))
                    
                # Atualiza o lado
                self.side = side
                
    # Função chamada a cada ordem preenchida
    def fill(self, id, instrument, price, quantity, status):
        print(f'Order Filled: {instrument} {quantity}@{price}')

print(evaluateHist(MonkeyTrade(), {'IBOV': '^BVSP.csv'}))
```

Exemplo de como pegar dados e salvar em um arquivo:

```python
import yahoofinancials as yf
ticker = 'F'

start_date = '2019-03-23'
end_date =  '2021-03-23'

data = yf.YahooFinancials(ticker).get_historical_price_data(start_date, end_date, 'daily')

# Ler os dados do JSON
raw = pd.DataFrame(data[ticker]['prices']).dropna()
# Converter a data para o tipo correto datetime
raw['formatted_date'] = pd.to_datetime(raw['formatted_date'])
# Indica a data como o índice de cada linha
raw = raw.set_index('formatted_date')
# Removendo as colunas que não interessam
df = raw.iloc[:,1:]

# Acertando a ordem das colunas
df = df.reindex(columns=['open', 'high', 'low', 'close', 'adjclose', 'volume'])
# Salvando o CSV
df.to_csv('F.csv')
```

E testar mais uma estratégia dummy (muito famosa em cassinos):

```python
class Martingale(Strategy):

    def __init__(self):
        self.trades = []
        self.size = 0
        self.average = 0

    def receive(self, event):

        price = event.price[3] # Captura o preço atual vindo do evento

        if self.size == 0: # Se está zerado
            self.submit(self._id, Order(event.instrument, Order.B, 1, 0))
            self.size = 1
            self.average = price
            # print("{0} {1} {2}".format(price, self.average, self.size))

        elif (price/self.average - 1) <= -0.03: # Se está caindo 3% do preço médio
            
            # dobra a posição
            self.submit(self._id, Order(event.instrument,  Order.B, self.size, 0))
            self.size *= 2
            # novo médio
            self.average += price
            self.average /= 2
            print("Price:{0} Average:{1} Position Quantity:{2}".format(price, self.average, self.size))
    
print(evaluateHist(Martingale(), {'F': 'F.csv'}))
```
