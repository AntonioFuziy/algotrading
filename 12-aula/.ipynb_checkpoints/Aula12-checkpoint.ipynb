### Value At Risk

Como visto, o Var pode ser medido de várias formas, sendo as mais comuns:

1. VaR paramétrico de uma carteira
1. VaR via simulação histórica de uma carteira
1. VaR via simulação Monte Carlo de uma carteira

Para exemplificar, vamos realizar medições usando apenas 1 ação

#### VaR Paramétrico

Consiste em aproximar a série de retornos por uma normal e extrair o percentil desejado do modelo.

Por exemplo: Vamos extrair o VaR 1% de PETR3 dos últimos 2 anos

````python
# VaR de 1 dia 99% confidence

# Considere que foram compradas R$ 10 mil em ações de PETR3
pos = 10000

# Calcula-se o retorno e remove os pontos faltantes
ret = df['PETR3.SA'].pct_change().dropna()

# 1-99% = 1% de alpha de uma normal padrão
z = norm.ppf(0.01, ret.mean(), ret.std())

# Value at Risk
var = pos*z

print('Total position: {}'.format(pos))
print('Value at Risk 1%: {}'.format(var))
````

Concluindo: Tendo uma carteira com cerca de 10 mil reais de PETR3, há 1% de chance de *perder* cerca de 800 reais em 1 dia segundo o cálculo do VaR paramétrico.

#### VaR Histórico

Para corrigir o problema do modelo acima, vamos usar a própria série de dados para calcular os percentis:

````python
# VaR Histórico de 1 dia 99% confidence

# Considere que foram compradas 10 mil em ações de PETR3
pos = 10000

# Calcula-se o retorno e remove os pontos faltantes
ret = df['PETR3.SA'].pct_change().dropna()

# 1-99% = 1% quantil dos DADOS
var = pos*ret.quantile(0.01)

print('Total position: {}'.format(pos))
print('Value at Risk 1%: {}'.format(var))
````

Note a discrepância de valores entre o paramétrico e o histórico.

#### VaR via Simulação de Monte Carlo

Consiste em construir a série, baseando-se na distribuição de dados passados e projetando para vários períodos futuros.

````python
# Calcula-se o retorno e remove os pontos faltantes
ret = df['PETR3.SA'].pct_change().dropna()

# Média dos Retornos
avg = ret.mean()

# Desvio padrão dos retornos
dp = ret.std()

# Dias de Projeção:
days = 5
# Quantidade de amostras
size = 1000

# Criando a matriz das amostras
pos = 10000 * np.ones((days+1, size))

# Aplicando a simulação
for i in range(1, days+1):
    pos[i,:] = np.multiply(pos[i-1,:], 1+norm.rvs(avg, dp, size=size))

plt.plot(pos);

print('VaR 1% {} dias via Simulação: {}'.format(days, np.quantile(pos[-1,:]-pos[0,:], 0.01)))
````

Note que quando usamos o número de dias = 1, o valor se aproxima do VaR Paramétrico!

#### Agora a Carteira!

Quando temos uma carteira de ativos, com certa alocação, podemos medir o VaR como nos moldes anteriores. CONTUDO deve-se levar em consideração a variação conjunta dos ativos (aka Covariância, e por tabela, a correlação).

Imagine que você está comprado em um ativo e vendido em um outro muito correlacionado ao primeiro. É factível pensar que você perderia ao extremo nas duas pontas simultaneamente?

Portanto, vamos realizar a mesma simulação anterior, mas com um modelo normal multivariado:

````python
from scipy.stats import multivariate_normal

# número de amostras
size = 10000

# Criando o vetor de posições
pos = np.array([[62520, 62520]]*size)

# Sinal
signal = np.array([1, 1])

# Vetor de posição da carteira de cada simulação
aloc = pos.sum(axis=1)

# Calcula-se o retorno e remove os pontos faltantes
ret = df[['PETR3.SA','PETR4.SA']].pct_change().dropna()

avg = ret.mean()
cov = ret.cov().values

# Projeção para uma semana:
days = 1

for i in range(days):
    pos = pos + pos * (multivariate_normal.rvs(avg, cov, size=size) * signal)

print('Alocação inicial: {}'.format(aloc[0]))
var = np.percentile(pos.sum(axis=1)-aloc, 1)
print('VaR 1% {} dias via Simulação: {}'.format(days, var))
````

### Ajustando o Modelo de Trading

* Modelo de sinal indica a direção (-1, 0, 1)
* Modelo de alocação balanceia os pesos conforme critério específico (Por exemplo max Sharpe Ratio)

Com os pesos na mão, falta agora decidir quanto de dinheiro será colocado. 
Para tal vamos fazer a conta inversa, por exemplo:

O limite de VaR de um 1 dia de uma Mesa é de R$10 mil

Dado os pesos da carteira, qual é a alocação que ele pode fazer para se manter dentro do VaR Esperado?

Solução: alocar um valor ponderado pelo modelo de alocação e medir o VaR. Achar o coeficiente que multiplica o VaR obtido para ajustar as alocações iniciais.

Por exemplo: Na simulação anterior, obter a alocação de PETR3 e PETR4 para atingir R$10 mil de VaR

````python
# Coeficiente
m = 10000 / var

# Alocação inicial era de 5 mil cada
print(m*5000)
````

Agora é só simular novamente!
Não esqueça que é aleatório, portanto não deve colocar no limite exato do VaR, mas algo próximo de limite (lembrando que dinheiro parado também é loss)

## Amarrando tudo

Vamos reviver um modelo de arbitragem estatística com co-integração, mas agora com múltiplos assets.

O índice IBOV é formado pelas 84 ações de maior volume na B3 e rebalanceado a cada 4 meses.

http://www.b3.com.br/pt_br/market-data-e-indices/indices/indices-amplos/indice-ibovespa-ibovespa-composicao-da-carteira.htm

Tecnicamente conseguimos reproduzir fielmente o índice se montarmos uma posição com os pesos sugeridos. 

IBOV não é um instrumento operável diretamente, sendo necessário de outras formas como:

* ETF
* Futuro
* Opções

Logo, bem teoricamente, teríamos uma forma de realizar uma arbitragem entre o índice e as ações que o compõem. Contudo operacionalmente isso é inviável por diversos motivos:
* Tratamento de muitas ordens simultâneas
* Problemas de liquidez
* Custos Operacionais
* etc

Portanto uma possível abordagem é realizar uma arbitragem **estatística**, usando apenas algumas ações com maior peso no índice. Para representar o índice, usaremos o ETF BOVA11, que replica o índice.

Usando as 5 ações mais pesadas e o modelo da aula de co-integração, temos:

$$ y_i=\omega_1x_i^1+\omega_2x_i^2+\omega_3x_i^3+\omega_4x_i^4+\omega_5x_i^5+\epsilon_i$$

Onde:
* $\omega$ é o peso da ação
* $x$ é o preço de cada ação
* $\epsilon$ é o resíduo

Vamos utilizar o mesmo código da aula:

````python
import statsmodels.api as sm

y = df['BOVA11.SA']
x = df[tickers[1:6]]

model = sm.OLS(y,x)

results = model.fit()

results.summary()
````

Finalmente, o backtesting

Rolling window: 2 anos de dados, sendo 1 ano para a co-integração e mais 6 meses para observar a banda

O código abaixo é uma junção da aula 10 com o conteúdo acima

````python
class CointMulti(Strategy):

    def __init__(self, tickers, var):
        
        # tickers, o primeiro é sempre o y da regressão
        self.tickers = tickers
        
        # Inicializando os dicionários
        self.prices = {}
        self.quantities = {}
        for ticker in tickers:
            self.prices[ticker] = []
            self.quantities[ticker] = 0
        
        # Limite de VaR
        self.var = var
        
        # Quantidade de dados para co-integração
        self.tam = 252
        # Quantidade de dados para as bandas dos resíduos
        self.window = 126
        
        self.side = 0
        
        self.indicator = []
        
        self.init_capital = 600000
        self.avail_capital = self.init_capital
        
    def receive(self, event):
        
        # Armazenando os preços:
        if event.instrument in self.tickers:
            self.prices[event.instrument].append(event.price[3])
        
        # Verificando a Sincronia
        sync = True
        for prices in self.prices.values():
            if len(prices) != len(self.prices[self.tickers[0]]):
                sync = False

        # Verificando o tamanho
        if sync and len(self.prices[self.tickers[0]]) >= self.tam:
            
            
            # Montagem
            y = pd.Series(self.prices[self.tickers[0]][-self.tam:])
            x = pd.DataFrame()
            for ticker in self.tickers[1:]:
                x[ticker] = self.prices[ticker][-self.tam:]
            
            # Regressão
            model = sm.OLS(y,x)
            results = model.fit()
            resid = list(results.resid)
            
            self.indicator.append(resid[-1])

            # Signal:
            signal = self.side
            if signal < 0 and resid[-1] < 0: # Condição para zerar
                signal = 0
            elif signal > 0 and resid[-1] > 0: # Condição para zerar
                signal = 0
                
            elif len(self.indicator) > self.tam: # Condição para entrar
                
                # Usando uma espécie de Bollinger Bands, a média será zero por construção
                if resid[-1] < -0.5*np.std(self.indicator[-self.window:]):
                    signal = 1
                elif resid[-1] > 0.5*np.std(self.indicator[-self.window:]):
                    signal = -1
                
            
            # Alocação:
            if self.side != signal:
                
                # Zeragem
                for ticker in self.tickers:
                    
                    if self.quantities[ticker] > 0:
                        self.submit(self.id, Order(ticker, Order.S, self.quantities[ticker], 0))
                    elif self.quantities[ticker] < 0:
                        self.submit(self.id, Order(ticker, Order.B, self.quantities[ticker], 0))                        
                
                # Separação dos preços
                prices = pd.DataFrame()
                for ticker in self.tickers:
                    prices[ticker] = self.prices[ticker]
                
                # Calcular média e covariância
                ret = prices.pct_change().dropna()
                avg = ret.mean()
                cov = ret.cov().values
                
                # Montar o vetor de posições
                pos = [self.prices[tickers[0]][-1]] # y
                sig = [-signal]
                for i in range(len(results.params)):
                    pos.append(np.abs(results.params[i])*self.prices[tickers[i+1]][-1])
                    if results.params[i] >= 0:
                        sig.append(signal)
                    else:
                        sig.append(-signal)
                        
                pos = np.array(pos)
                sig = np.array(sig)
                
                # Calcular o VaR com alocação inicial (pesos da regressão)
                size = 1000
                aloc = pos.sum()
                pos = pos + pos * (multivariate_normal.rvs(avg, cov, size=size) * sig)
                var = np.percentile(pos.sum(axis=1) -aloc, 1)
                
                # Ajustado para o limite de VaR
                if np.abs(var) < 1: # se o var for muito baixo, limita-se à liquidez
                    coef = 10000
                else:
                    coef = self.var / var
                
                # Calculando as quantidades
                # y
                self.quantities[tickers[0]] = -signal*coef
                # x
                for i in range(len(self.tickers[1:])):
                    self.quantities[tickers[i+1]] = coef*signal*results.params[i]
               
                # Montagem da posição
                for ticker in self.tickers:
                    
                    if self.quantities[ticker] > 0:
                        self.submit(self.id, Order(ticker, Order.B, self.quantities[ticker], 0))
                    elif self.quantities[ticker] < 0:
                        self.submit(self.id, Order(ticker, Order.SS, self.quantities[ticker], 0))                                            
                    
                # Atualiza o sinal
                self.side = signal

    def fill(self, id, instrument, price, quantity, status):
        super().fill(id, instrument, price, quantity, status)
        
        # Imprimindo o preenchimento das ordens
        if quantity != 0:
            print('Fill: {0} {1}@{2}'.format(instrument, quantity, price))
````

````python
model = CointMulti(tickers[:6], 10000)

files = {}
for ticker in tickers[:6]:
    files[ticker] = '{}.csv'.format(ticker)
    
print(evaluateHist(model, files))
````
