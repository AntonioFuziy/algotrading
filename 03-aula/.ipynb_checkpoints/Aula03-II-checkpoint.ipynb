Agora vamos baixar dados de fechamento (**Historical Data**) de 6 meses diretamente do site

```python
import yahoofinancials as yf
ticker = '^BVSP'

start_date = '2020-02-23'
end_date =  '2023-02-23'

data = yf.YahooFinancials(ticker).get_historical_price_data(start_date, end_date, 'daily')
```

Agora que sabemos como separar o que interessa, vamos começar a montar nosso dataframe

```python
# Ler os dados do JSON
ibov_raw = pd.DataFrame(data[ticker]['prices']).dropna()
# Converter a data para o tipo correto datetime
ibov_raw['formatted_date'] = pd.to_datetime(ibov_raw['formatted_date'])
# Indica a data como o índice de cada linha
ibov_raw = ibov_raw.set_index('formatted_date')
# Visualiza o resultado
ibov_raw.head()
```

Vamos remover as colunas que não interessam muito, mantendo o padrão OHLC (Open, High, Low, Close) para de fechamento, mais o volume

```python
ibov = ibov_raw.iloc[:,1:-1]
ibov.head()
```

Voltando aos dados diários, vamos dar uma olhadinha brevemente nos retornos

```python
ibov_ret = ibov["close"].pct_change().dropna()*100
ibov_ret.head(10)
```

Explorando os dados

```python
# Medidas Resumo
ibov_ret.describe()
```

```python
# Histograma
ibov_ret.plot.hist(density=True);
```

```python
# Histograma
import scipy.stats as stats
stats.probplot(ibov_ret, dist="norm", plot=plt);
```

```python
# Boxplot
ibov_ret.plot(kind='box');
```

Calculando o log-retorno

```python
ibov_logr = ibov.close.apply(np.log).diff().dropna()*100
ibov_logr.head()
```

Apenas verificando se log-retorno e retorno são muito diferentes para esses dados

```python
(ibov_ret - ibov_logr).plot();
```

### Simulando uma Estratégia de Trading

Vamos começar olhando como interagir com os dados:

```python
# Percorrendo todos os preços de fechamento
for price in ibov.close:
    print(price)
```

Imagine agora que você comprou 1 instrumento no próprio preço no primeiro ponto:

```python
# Primeiro ponto:
print(ibov.close[0])
```

```python
# Montando a alocação

notional = ibov.close[0]
print(notional)
```

Agora imagine que vendeu no último dia:

```python
# Resultado em termos monetários
result = ibov.close[-1] - notional
print(result)
```

```python
# Em percentual (retorno):
print(f'{result/notional*100:0.2f}%')
```

Resultado: Se comprou IBOV 3 anos atrás e vendeu no dia 23/Fev/2023, perdeu cerca de R$ 1,346.00 por instrumento.

PnL (Profit & Loss): 1,27%

Parece um bom negócio! Mas será que é tão simples?

___

Para contar toda a história do trade, queremos saber quanto ela rendeu ao longo do tempo (marcação a mercado - MTM), preciso analisar o quanto seria o meu resultado a cada dia que passa:

```python
# Comprei no primeiro ponto
notional = ibov.close[0]
for price in ibov.close:
    # Para cada ponto, calculo meu resultado se fosse vender
    print(price - notional)
```

Não foi muito útil, vamos gerar o gráfico:

```python
# Comprei no primeiro ponto
notional = ibov.close[0]

result = []
for price in ibov.close:
    # Para cada ponto, calculo meu resultado se fosse vender e guardo em uma lista
    result.append(price - notional)
    
pd.Series(result).plot();
```

Agora sim, e não pareceu tão bonito quanto o resultado final. Qual foi o pior momento?

```python
min(result)
```

```python
print(f'{min(result)/notional*100:0.2f}%')
```

Esse resultado é o Max Drawdown, ou seja, o maior prejuízo ao longo do tempo. Você consegue calcular o Max Profit?

Outra medida a ser observada é a taxa de acerto, mas como só foi feito 1 operação, essa medida não faz muito sentido ainda.
___

Para termos uma maior acurácia, vamos começar a trabalhar com fluxo de caixa:
 * Toda vez que compra, dinheiro sai do bolso
 * Toda vez que vende, dinheiro entra
 
Vamos também montar um gráfico mais interativo

```python
fig, ax = plt.subplots(1, 1, figsize=(10,5));

cash = 0 # indica quanto dinheiro tem no bolso
position = 0 # indica a posição atual. 0 é zerado
result = [] # lista para montar o gráfico

fim = 0

for price in ibov.close:
    
    # Se a posição está zerada, compra
    if position == 0 and fim == 0:
        cash -= 1 * price # Fluxo de caixa negativo
        position = 1 # Indica que está comprado em 1
    
    result.append(cash + 1 * price * position) # Mark To Market!
   
    # Fazendo o plot a cada iteração:
    ax.plot(result, color='blue')    
    display.clear_output(wait=True)
    display.display(fig)
    
ax.plot(result, color='blue');
display.clear_output(wait=True)
```

Vamos fazer uma pequena modificação: vender quando atingir 2% de lucro. Para isso precisamos armazena o preço de compra para poder comparar com o preço atual:

```python
fig, ax = plt.subplots(1, 1, figsize=(10,5));

cash = 0 # indica quanto dinheiro tem no bolso
position = 0 # indica a posição atual. 0 é zerado
result = [] # lista para montar o gráfico
buy_price = 0
sell_price = 0

fim = 0

for price in ibov.close:
    
    # Se a posição está zerada, compra
    if position == 0 and fim == 0:
        cash -= 1 * price # Fluxo de caixa negativo
        position = 1 # Indica que está comprado em 1
        buy_price = price        
        
    # Critério de Venda:
    if position == 1 and price > buy_price * 1.02:
        cash += 1 * price
        position = 0
        sell_price = price
        fim = 1
    
    result.append(cash + 1 * price * position) # Mark To Market!
   
    # Fazendo o plot a cada iteração:
    ax.plot(result, color='blue')    
    display.clear_output(wait=True)
    display.display(fig)
    
ax.plot(result, color='blue');
display.clear_output(wait=True)
```
