```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
import datetime
import random
import yahoofinancials as yf
import pandas_ta as ta

# Para fazer o backtesting
from backtesting_v2 import evaluateHist, evaluateIntr, Strategy, Order

plt.style.use("dark_background")
```

```python
# instalando o pandas-ta
ticker = 'AAPL'
df = pd.read_csv('{}.csv'.format(ticker))
df.head()
```

```python
class CMOStrategy(Strategy):

    def __init__(self):
        self.period = 14
        self.prices = []
        self.cmo_indicator = []
        
        
        # Position side (+1 for long, -1 for short, 0 for flat)
        self.side = 0
        self.init_capital = 100
        self.avail_capital = self.init_capital


    def receive(self, event):
        signal = 0
        price = event.price[3] # Current price from event
        self.prices.append(price)

        if len(self.prices) >= self.period + 1:
            close = pd.DataFrame({'close': self.prices})
            cmo = close.ta.cmo(length=self.period)
            self.cmo_indicator.append([cmo.iloc[-1]])
            
            cmo_value = cmo.iloc[-1]
            
            if cmo_value >= 50: 
                signal = 1
            elif cmo_value <= -50:
                signal = -1
                
            if self.side != signal:
                if self.side == 1:
                    self.submit(self.id, Order(event.instrument, Order.S, self.side, 0))
                elif self.side == -1:
                    self.submit(self.id, Order(event.instrument, Order.B, -self.side, 0))

                if signal == 1:
                    self.submit(self.id, Order(event.instrument, Order.B, signal, 0))
                else:
                    self.submit(self.id, Order(event.instrument, Order.SS, -signal, 0))

                self.side = signal



cmo_strategy = CMOStrategy()
print(evaluateHist(cmo_strategy, {'AAPL': 'AAPL.csv'}))
```

```python
fig, ax = plt.subplots(3, 1, figsize=(15,10));

ax[0].plot([day[0] for day in cmo_strategy._days.values()][-len(cmo_strategy.cmo_indicator):])
ax[0].legend(['Capital']);
ax[1].plot(cmo_strategy.prices[-len(cmo_strategy.cmo_indicator):], color='white');
ax[1].legend(['Price']);
ax[2].plot(cmo_strategy.cmo_indicator);
ax[2].legend(['CMO', 'signal']);
```

```python
class PPOStrategy(Strategy):

    def __init__(self):
        self.ppo_fast = 12
        self.ppo_slow = 26
        self.ppo_signal = 9
        self.ppo_scalar = 100
        
        # Lists to store price data and indicators
        self.prices = []
        self.ppo_indicator = []
        
        # Position side (+1 for long, -1 for short, 0 for flat)
        self.side = 0

    def receive(self, event):
        price = event.price[3] # Current price from event
        
        # Add current price to price list
        self.prices.append(price)

        if len(self.prices) >= self.ppo_slow + self.ppo_signal:
            close = pd.DataFrame({'close': self.prices})
            
            ppo = close.ta.ppo(fast=self.ppo_fast, slow=self.ppo_slow, signal=self.ppo_signal, scalar=self.ppo_scalar)
            
            self.ppo_indicator.append([ppo.iloc[-1, 0], ppo.iloc[-1, 2]])
            
            ppo_signal = (ppo.iloc[-1, 0] > ppo.iloc[-1, 2]) * 2 - 1

            if self.side != ppo_signal:
                if self.side == 1:
                    self.submit(self.id, Order(event.instrument, Order.S, self.side, 0))
                elif self.side == -1:
                    self.submit(self.id, Order(event.instrument, Order.B, -self.side, 0))

                if ppo_signal == 1:
                    self.submit(self.id, Order(event.instrument, Order.B, ppo_signal, 0))
                else:
                    self.submit(self.id, Order(event.instrument, Order.SS, -ppo_signal, 0))

                self.side = ppo_signal
                

ppo_strategy = PPOStrategy()
print(evaluateHist(ppo_strategy, {'AAPL': 'AAPL.csv'}))
```

```python
fig, ax = plt.subplots(3, 1, figsize=(15,10));

ax[0].plot([day[0] for day in ppo_strategy._days.values()][-len(ppo_strategy.ppo_indicator):])
ax[0].legend(['Capital']);
ax[1].plot(ppo_strategy.prices[-len(ppo_strategy.ppo_indicator):], color='white');
ax[1].legend(['Price']);
ax[2].plot(ppo_strategy.ppo_indicator);
ax[2].legend(['PPO', 'signal']);
```

```python
class Bollinger(Strategy):

    def __init__(self):
        
        # Quantidade de dias de cada mdia mvel
        self.period = 14
        
        # Nmero de desvios da banda
        self.m = 1 # testar 1.5
        
        # Lista para guardar os preos e o indicador
        self.prices = []
        self.indicator = []
    
        # Side da estratgia (+1 Compra/-1 Vende)
        self.side = 0
        
        self.init_capital = 100
        self.avail_capital = self.init_capital


    def receive(self, event):
        
        price = event.price[3] # Captura o preo atual vindo do evento
        
        # Montando a srie de preos:
        self.prices.append(price)
        
        # Se a quantidade de preos observados for superior a janela
        if len(self.prices) >= self.period + 1:
            
            # Monta a lista de preos em um dataframe
            close = pd.DataFrame({'close': self.prices})
            
            bbands = close.ta.bbands(lenght = self.period, std = self.m)            
            
            down = bbands.iloc[-1, 0]
            avg = bbands.iloc[-1, 1]
            up = bbands.iloc[-1, 2]
            
            # Guarda os dados para o grfico
            self.indicator.append([up, avg, down])
                                    
            # Se o preo estiver abaixo da banda inferior compra, seno vende
            if price < down:
                signal = -1 
            elif price > up:
                signal = 1
            elif price > avg and self.side == -1:
                signal = 0
            elif price < avg and self.side == 1:
                signal = 0
            else:
                signal = self.side
            
            # Ajustando a posio
            if self.side != signal:
                
                if self.side == 1:
                    self.submit(self.id, Order(event.instrument, Order.S, self.side, 0))
                elif self.side == -1:
                    self.submit(self.id, Order(event.instrument, Order.B, -self.side, 0))
                
                if signal == 1:
                    self.submit(self.id, Order(event.instrument, Order.B, signal, 0))
                else:
                    self.submit(self.id, Order(event.instrument, Order.SS, -signal, 0))
                    
                self.side = signal

modelBollinger = Bollinger()
print(evaluateHist(modelBollinger, {'AAPL': 'AAPL.csv'}))
```

```python
# Plot do grfico
fig, ax = plt.subplots(2, 1, figsize=(15,15));

shift = -len(modelBollinger.indicator)
ax[0].plot([day[0] for day in modelBollinger._days.values()][shift:100+shift])
ax[0].legend(['Capital']);
ax[1].plot(modelBollinger.prices[shift:100+shift], color='white');
ax[1].plot(modelBollinger.indicator[:100]);
ax[1].legend(['Price', 'Upper band', 'Average', 'Lower band']);
```
